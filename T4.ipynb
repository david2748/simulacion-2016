```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
va=1664525
vb=1013904223
vm=2**32
def Aleatorio():
    u=[]
    x=24242
    for i in range(1000):
        x=(va*x+vb)%vm
        u.append(float(x)/vm)
    return u
def Uniforme():
    p=Aleatorio()
    a=2;
    b=6;
    F=[]
    for i in range(1000):
        F.append(a+p[i]*(b-a))
    plt.hist(F, 25, normed=True)
    uniformeTeorico(a,b,1000)
    plt.show()
    
def uniformeTeorico(a,b,escala):
    x=np.linspace(a,b,escala)
    f=[]
    for i in range(len(x)):
        f.append(1/(b-a))
    plt.plot(x,f)

def Logistica():
    p=Aleatorio()
    m=4
    s=3
    F=[]
    for i in range(1000):
        F.append(m+(s*math.log(p[i]/float(1-p[i]))))
    plt.hist(F, 25, normed=True)
    x=np.linspace(-15,20,1000)
    f=[]
    for i in range(len(x)):
        f.append((math.exp(-(x[i]-m)/float(s)))/(s*(1+math.exp(-(x[i]-m)/float(s)))**2))
    plt.plot(x,f)
    plt.show()


def Weibull():
    p=Aleatorio()
    l=1
    k=3
    F=[]
    for i in range(1000):
        F.append((l*(-math.log(1-p[i])))**(1/float(k)))
    plt.hist(F, 25, normed=True)
    x=np.linspace(0.2,1.6,1000)
    f=[]
    for i in range(len(x)):
        f.append((k/float(l))*((x[i]/float(l))**(k-1))*(math.exp((-(x[i]/float(l))**k))))
    plt.plot(x,f)
    plt.show()
print("Uniforme")
Uniforme()
print("logistica")
Logistica()
print("Weibull")
Weibull()
```
